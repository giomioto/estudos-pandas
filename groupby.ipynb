```python
import pandas as pd
```

```python
df = pd.read_csv("bank.csv", encoding = "UTF-8", sep =";")
```

```python
df.age.mean()
```

```python
df[(df.marital == "single")].age.mean()
```

```python
#groupby

df[["marital", "education","age"]].groupby(["marital", "education"]).mean()

```

```python
colunas = ["balance", "marital"]
df[((df.age >= 30) & (df.age <= 50)) & (df.marital.isin(["married","divorced"]))][colunas].groupby(["marital"]).mean()
```

```python
df[((df.age >= 30) & (df.age <= 50)) & (df.marital.isin(["single"]))][["job", "marital"]].groupby(["job"]).count()
```

```python
df[((df.age >= 30) & (df.age <= 50)) & (df.marital.isin(["single"]))][["job","education", "marital"]].groupby(["job", "education"]).count()
```

```python
df[((df.age >= 30) & (df.age <= 50)) & (df.marital.isin(["single"]))][["job","education", "marital"]].groupby(["job", "education"]).value_counts()
```
