```python
import pandas as pd
```

```python
microdadosEnem = pd.read_csv("microdados_enem_2015\DADOS\MICRODADOS_ENEM_2015.csv", encoding = "ISO-8859-1", sep = ";")
```

```python
microdadosEnem.columns.values
```

```python
microdadosEnemSelecionados = microdadosEnem.filter(items = ['CO_MUNICIPIO_PROVA','NO_MUNICIPIO_PROVA','CO_UF_PROVA','SG_UF_PROVA','NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
       'NU_NOTA_MT','NU_NOTA_REDACAO'])
```

```python
microdadosEnemSelecionados.dropna(how='any', inplace=True)
microdadosEnemSelecionados
```

```python
mediaEnemPorMunicipio = microdadosEnemSelecionados.groupby\
(['CO_MUNICIPIO_PROVA','NO_MUNICIPIO_PROVA','CO_UF_PROVA','SG_UF_PROVA'], as_index = False).mean()
```

```python
mediaEnemPorMunicipio
```

```python
mediaEnemPorMunicipio.to_csv("mediaEnemPorMunicipio")
```
